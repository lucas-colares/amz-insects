```python
from sahi import AutoDetectionModel
from sahi.utils.cv import read_image
from sahi.utils.file import download_from_url
from sahi.predict import get_prediction, get_sliced_prediction
from IPython.display import Image
import pandas as pd
import glob
import re
```

```python
import os
for x in list(range(1,6)):
  os.makedirs(os.path.join("IDs",str(x)+"-fold"))
```

```python
for u in list(range(4,6)):
  detection_model = AutoDetectionModel.from_pretrained(
      model_type="yolov8",
      model_path=os.path.join("model_training","runs","train"+str(u),"weights","best.pt"),
      confidence_threshold=0,
      device="cuda:0"
  )

  imgs = glob.glob("/content/originals/*.JPG") #Especificamos aqui o caminho onde estão nossas imagem para identificação
  import re
  import pandas as pd
  from tqdm.notebook import tqdm
  from PIL import Image

  for y in tqdm(imgs):
    result = get_sliced_prediction(
        y,
        detection_model,
        slice_height=500,
        slice_width=500,
        overlap_height_ratio=0.2,
        overlap_width_ratio=0.2)
    filename = y.split("/")[len(y.split("/"))-1]
    data_list = []
    for x in list(range(0,len(result.object_prediction_list))): #list(range(0,8)), ou seja, 0, 1, 2, 3, 4, 5, 6 e 7 serão os números que substituirão o "x" em cada rodada do for
      bbox = re.findall("\(.*?\)",str(result.object_prediction_list[x])) # "str" transforma a lista em uma string
      bbox = str(bbox).split(",")
      score = re.findall("value:.*?>",str(result.object_prediction_list[x])) # "str" transforma a lista em uma string
      score_final = re.sub("value: |>","",score[0])
      name = re.findall("name:.*?>",str(result.object_prediction_list[x]))  # "str" transforma a lista em uma string
      name_final = re.sub("name: |>","",name[0])
      data_list.append([re.sub(r"\(|np.float64|\)| |\[|\]|'", "", bbox[0]),re.sub(r"\(|np.float64|\)| |\[|\]|'", "", bbox[1]),re.sub(r"\(|np.float64|\)| |\[|\]|'", "", bbox[2]),re.sub(r"\(|np.float64|\)| |\[|\]|'", "", bbox[3]),score_final,name_final])
    df = pd.DataFrame(data_list, columns=['xmin','ymin','xmax','ymax','conf','class'])
    df.to_csv('/content/IDs/'+str(u)+'-fold/'+re.sub(".JPG",".csv",filename))
```
